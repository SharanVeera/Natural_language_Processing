**Stemming** is a text preprocessing technique in Natural Language Processing (NLP). Specifically, it is the process of reducing inflected form of a word to one so-called “stem,” or root form, also known as a “lemma” in linguistics.1 It is one of two primary methods—the other being lemmatization—that reduces inflectional variants within a text dataset to one morphological lexeme. In doing so, stemming aims to improve text processing in machine learning and information retrieval systems.

```python
## porter stemmer
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
words = ["eating","eats","writing","wrote","written","writes","healthy","finally","finalized","programms","programming"]
```

```python
for word in words:
  print(word+"---->"+stemming.stem(word))
```

```python
stemming.stem("congratulations")
```

**Regex stemming**, or Regular Expression Stemming, is a method of stemming that utilizes regular expressions to identify and remove common suffixes or prefixes from words, aiming to reduce them to their root form (stem). This approach is particularly useful for custom or domain-specific stemming tasks where predefined rules are necessary.

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|able$|e&',min=4)
reg_stemmer.stem("congratulations")
```

```python
reg_stemmer.stem("understandable")
```

**Snowball stemming Algorithm:** It reduces words to their base or root form (stem) by removing suffixes and common word endings. For instance, "running," "runs," and "ran" would all be stemmed to "run."

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')
for word in words:
  print(word+"--->"+stemmer.stem(word))
```

```python
##porter stemmer
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
## snowball stemmer
stemmer.stem("fairly"),stemmer.stem("sportingly")
```
